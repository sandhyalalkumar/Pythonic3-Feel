```python
import numpy as np
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
iris.data
```

```python
X = iris.data
Y = iris.target
```

```python
X_train = X[:100]
Y_train = Y[:100]
```

```python
X_validate = X[100:]
Y_validate = Y[100:] 
```

```python
from sklearn.neighbors import KNeighborsClassifier
neighbors = KNeighborsClassifier(n_neighbors=2, weights='uniform')
neighbors.fit(X_train, Y_train)
predicted = neighbors.predict(X_validate)
print(predicted)
```
