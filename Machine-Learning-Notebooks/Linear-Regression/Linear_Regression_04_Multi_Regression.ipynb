```python
X = [df.columns[:-1]]
Y = [df.columns[-1]]
print(X)
print(Y)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,random_state=0)
reg=LinearRegression()
reg.fit(xtrain,ytrain)
print(list(reg.predict(xtrain))[:5])
print(reg.score(xtest,ytest))
```
